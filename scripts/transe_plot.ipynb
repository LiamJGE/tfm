```python
import matplotlib.pyplot as plt

# Define the entities and relations
entities = {
    'A': (1, 2),
    'B': (3, 4),
    'C': (5, 0)
}

relations = {
    'r1': (2, 2),
    'r2': (2, -4)
}

# Calculate positions after relations
B_from_A_r1 = (entities['A'][0] + relations['r1'][0], entities['A'][1] + relations['r1'][1])
C_from_B_r2 = (entities['B'][0] + relations['r2'][0], entities['B'][1] + relations['r2'][1])

# Plot the entities
for entity, coords in entities.items():
    plt.scatter(*coords, label=entity, color='black', marker='x')
    plt.text(coords[0] + 0.1, coords[1], entity, fontsize=12)

# Plot the calculated positions
# plt.scatter(*B_from_A_r1, label='A + r1', color='blue', marker='x')
# plt.scatter(*C_from_B_r2, label='B + r2', color='red', marker='x')
# plt.text(B_from_A_r1[0] + 0.1, B_from_A_r1[1], 'A + r1', fontsize=12, color='blue')
# plt.text(C_from_B_r2[0] + 0.1, C_from_B_r2[1], 'B + r2', fontsize=12, color='red')

# Plot the relations as arrows
plt.arrow(entities['A'][0], entities['A'][1], relations['r1'][0], relations['r1'][1],
          head_width=0.05, head_length=0.1, fc='blue', ec='blue', label='r1')
plt.arrow(entities['B'][0], entities['B'][1], relations['r2'][0], relations['r2'][1],
          head_width=0.05, head_length=0.1, fc='red', ec='red', label='r2')

# Add labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
# plt.title('TransE Example in 2D Space')
plt.legend()

# plt.grid()
plt.show()
```
